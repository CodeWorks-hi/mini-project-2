```python
# ---- 데이터 전처리 ----
# 필요 없는 컬럼 제거

df2 = df2.rename(columns={"계정항목": "Country"})  # 국가 컬럼명 변경

# 데이터 변환: 국가별 성장률을 하나의 컬럼으로 정리
df2_melted = df2.melt(id_vars=["Country"], var_name="Date", value_name="Growth Rate")

# ---- 대표 국가 3개 선택 ----
selected_countries = ["한국", "미국", "중국"]
df2_selected = df2_melted[df2_melted["Country"].isin(selected_countries)]
```

```python
# ---- 시각화 ----
plt.figure(figsize=(12, 6))
sns.lineplot(data=df2_selected, x="Date", y="Growth Rate", hue="Country", marker="o")

plt.xticks(rotation=45)  # X축 눈금 회전
plt.xlabel("연도")
plt.ylabel("경제성장률 (%)")
plt.title("연도별 대표 3개국 경제성장률 추이")
plt.legend(title="국가", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```
