```python
# BSI코드에서 신규수주실적과 생산설비실적만 불러오기
linkage_analysis=df[df['BSI코드'].str.contains('신규수주실적BSI|생산설비실적BSI')]
```

```python
linkage_analysis
```

```python
linkage_analysis=linkage_analysis.T.drop('BSI코드')
```

```python
linkage_analysis.columns=['신규수주실적BSI','생산설비실적BSI']
```

```python
linkage_analysis.reset_index(inplace=True)
```

```python
linkage_analysis['신규수주실적BSI'] = linkage_analysis['신규수주실적BSI'].astype(float)
linkage_analysis['생산설비실적BSI'] = linkage_analysis['생산설비실적BSI'].astype(float)
```

```python
linkage_analysis = linkage_analysis.dropna(subset=['신규수주실적BSI', '생산설비실적BSI'])
```

```python
linkage_analysis
```

```python
# 산점도 그리기
plt.figure(figsize=(10, 6))
sb.scatterplot(x='신규수주실적BSI', y='생산설비실적BSI', data=linkage_analysis)

# 제목과 레이블 추가
plt.title('신규수주실적과 생산설비실적의 관계', fontsize=15)
plt.xlabel('신규수주실적BSI', fontsize=12)
plt.ylabel('생산설비실적BSI', fontsize=12)

# 추세선 추가
sb.regplot(x='신규수주실적BSI', y='생산설비실적BSI', data=linkage_analysis, scatter=False, color='red')

# 상관계수 계산 및 표시
correlation = linkage_analysis['신규수주실적BSI'].corr(linkage_analysis['생산설비실적BSI'])
plt.text(0.05, 0.95, f'상관계수: {correlation:.2f}', transform=plt.gca().transAxes, fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# 신규수주실적과 생산설비싱적 간에 강한 양의 상관관계가 관찰됨.
# 이는 신규수주가 중가할때 생산설비도 함께 증가하는 경향을 보임
# 점들이 우상향하는 패턴을 보이며 신규수주증가가 생산설비확충 또는 가동 증가로 아어지는 것으로 해석
# 신규수주증가가 생산설비실적에 등정적인 영향을 미치며, 기업들이 수요 변화에 따라 생산능력을 조정하고 있음을 시사함
# 설비투자는 기업가의 경기판단부터 설비 가동까지 여러 단계를 거칩니다. 
# 이 과정에서 신규 수주 증가는 설비투자 계획 수립으로 이어지고, 결과적으로 생산설비실적 증가로 나타남
# 신규 수주 증가 시 기업들은 생산능력을 확대하기 위해 설비투자를 늘리는 경향이 있음
# 신규 수주 증가는 미래 수요 증가를 의미하므로, 기업들은 이에 대응하여 생산설비를 확충하고 있음
```
