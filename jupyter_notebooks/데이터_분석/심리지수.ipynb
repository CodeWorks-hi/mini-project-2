```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
```

```python
df=pd.read_csv('/Users/yeonu/yeonu/git/mini-project-2/extra_data/processed/소비 심리 관련/경제심리지수.csv')
```

```python
df
```

```python
df2=df.T
```

```python
df2.reset_index(inplace=True)
```

```python
df2.columns=['date','경제심리지수(원계열)','경제심리지수(순환변동치)']
```

```python
df2.reset_index(inplace=True)
```

```python
df2=df2.drop([0,1,2,3])
```

```python
df2['date']=pd.to_datetime(df2['date'])
```

```python
df2.drop('index',axis=1,inplace=True)
```

```python
df2
```

```python
# 그래프 시각화
plt.figure(figsize=(10,5))
plt.plot(df2['date'],df2['경제심리지수(원계열)'],label='경제심리지수(원계열)')
plt.plot(df2['date'],df2['경제심리지수(순환변동치)'],label='경제심리지수(순환변동치)')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# 뉴스심리지수와 경제심리지수(순환변동치) 비교
plt.figure(figsize=(10,5))
plt.plot(df2['date'],df2['경제심리지수(순환변동치)'],label='경제심리지수(순환변동치)')
plt.plot(news_df['date'],news_df['뉴스심리지수'],label='뉴스심리지수')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
