```python
df.replace("-", np.nan, inplace=True)
```

```python
import pandas as pd
import numpy as np

for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='ignore')  # 오류 발생 시 object 유지

# 숫자로 변환된 컬럼만 골라내기
numeric_cols = df.columns[df.dtypes.apply(lambda x: np.issubdtype(x, np.number))]

# 각 행의 결측치 및 0을 최솟값~최댓값 사이 임의 값으로 대체
def fill_row_with_random(row):
    values = row[numeric_cols]
    values = values.replace(0, np.nan)
    non_missing = values.dropna()

    if len(non_missing) > 0:
        min_val = non_missing.mean()
        max_val = non_missing.max()

        if min_val == max_val:
            filled = values.apply(lambda x: min_val if pd.isna(x) else x).round(0)
        else:
            filled = values.apply(lambda x: np.random.uniform(min_val, max_val) if pd.isna(x) else x).round(0)

        row[numeric_cols] = filled
    return row

df = df.apply(fill_row_with_random, axis=1)
```

```python
df
```

```python
df_numeric = df.copy()
for col in df.columns:
    df_numeric[col] = pd.to_numeric(df[col], errors='coerce')

# 2024-02의 결측치만 각 행의 최솟값~최댓값 사이 임의 값으로 대체
def fill_random_in_row(row):
    if pd.isna(row['2024-02']):
        numeric_values = row.dropna()
        min_val = numeric_values.min()
        max_val = numeric_values.max()
        row['2024-02'] = np.random.uniform(min_val, max_val)
    return row

df_filled = df_numeric.apply(fill_random_in_row, axis=1).round()
```

```python
df.iloc[:, 3:] = df_filled.iloc[:, 3:]
```
