```python
#  melt로 월 데이터를 세로로 정리
월컬럼 = [f"{i}월" for i in range(1, 13)]
월매핑 = {f"{i}월": i for i in range(1, 13)}
```

```python
df_zs_long = df_zs.melt(id_vars=['차량 구분', '국가명', '연도'], value_vars=월컬럼, var_name='월', value_name='수출량')
df_cp_long = df_cp.melt(id_vars=['차종', '차량 구분', '거래 유형', '연도'], value_vars=월컬럼, var_name='월', value_name='판매량')
df_fp_long = df_fp.melt(id_vars=['공장명(국가)', '공장 코드', '차종', '연도'], value_vars=월컬럼, var_name='월', value_name='공장판매량')
df_lp_long = df_lp.melt(id_vars=['국가명', '공장명(국가)', '차종', '연도'],value_vars=월컬럼,var_name='월', value_name='현지판매량')
```

```python
# 1️.  수출량(df_zs_long) + 판매량(df_cp_long) 병합
df_merge1 = pd.merge(
    df_zs_long,
    df_cp_long,
    on=['차량 구분', '연도', '월'],
    how='outer'
)
print(" df_merge1 병합 완료:", df_merge1.shape)
```

```python
# 결측치 처리
for col in ['수출량', '판매량']:
    df_merge1[col] = pd.to_numeric(df_merge1[col], errors='coerce').fillna(0)

for col in ['국가명', '차종', '거래 유형']:
    if col in df_merge1.columns:
        df_merge1[col] = df_merge1[col].fillna(df_merge1[col].mode()[0])
```

```python
# 2️.  df_merge1 + 공장판매량(df_fp_long) 병합
df_merge2 = pd.merge(
    df_merge1,
    df_fp_long,
    on=['차종', '연도', '월'],
    how='outer'
)
print(" df_merge2 병합 완료:", df_merge2.shape)
```

```python
# 결측치 처리
df_merge2['공장판매량'] = pd.to_numeric(df_merge2['공장판매량'], errors='coerce').fillna(0)

for col in ['공장명(국가)', '공장 코드']:
    if col in df_merge2.columns:
        df_merge2[col] = df_merge2[col].fillna(df_merge2[col].mode()[0])
```

```python
# 3️. 위 결과 + df_lp_long (국가명 포함 버전 추천)
df_merge = pd.merge(
    df_merge2,
    df_lp_long,
    on=['차종', '공장명(국가)', '연도', '월'],
    how='outer'
)
print("최종 df_merge 병합 완료:", df_merge.shape)
```

```python
# 결측치 처리
df_merge['현지판매량'] = pd.to_numeric(df_merge['현지판매량'], errors='coerce').fillna(0)
```

```python

# 국가명 컬럼이 병합 중 두 개 생겼다면 처리
if '국가명_x' in df_merge.columns and '국가명_y' in df_merge.columns:
    df_merge['국가명'] = df_merge['국가명_x'].combine_first(df_merge['국가명_y'])
    df_merge.drop(columns=['국가명_x', '국가명_y'], inplace=True)

# 국가명 최빈값 보완
if '국가명' in df_merge.columns:
    df_merge['국가명'] = df_merge['국가명'].fillna(df_merge['국가명'].mode()[0])
```

```python
df_2024 = pd.read_csv("/Users/mi-yeonlim/Documents/Github/mini-project-2/data/raw/현대/해외공장판매실적_2024.csv")
```

```python
# 월별 판매량 컬럼을 숫자형으로 변환 (쉼표 제거 및 결측값 처리)
monthly_columns = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
for col in monthly_columns:
    df_2024[col] = pd.to_numeric(df_2024[col].replace({',': ''}, regex=True), errors='coerce')

# 각 공장의 총 판매량 계산 (월별 합계)
df_2024['Total Sales'] = df_2024[monthly_columns].sum(axis=1)

# 공장별 총 판매량 계산
sales_by_plant = df_2024.groupby('Plant')['Total Sales'].sum().sort_values(ascending=False)

# 시각화
plt.figure(figsize=(12, 6))
sales_by_plant.plot(kind='bar', color='skyblue')
plt.title('2024년 공장별 총 판매량')
plt.xlabel('공장')
plt.ylabel('판매량')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 주요 관찰 사항
1. EV9, EV5 등 전기차 모델의 도입 및 판매 증가
2. 지역별 특화 모델 존재 (예: 중국의 KX 시리즈)

```python
# 차종 키워드 기반 차량단가 설정
def get_unit_price(model):
    if pd.isna(model):
        return 22000000
    model = model.lower()
    if "ev" in model or "전기차" in model or "아이오닉" in model or "ev6" in model:
        return 40000000
    elif "suv" in model or "투싼" in model or "스포티지" in model:
        return 28000000
    else:
        return 22000000  # 기본 세단

df['차량단가'] = df['차종'].apply(get_unit_price)
```
