```python
import pandas as pd
```

```python
df = pd.read_csv("../../extra_data/processed/경제 성장 관련/국제 주요국 경제성장률(연도별).csv")
df
```

```python
target = ['한국', '호주', '칠레', '중국', '아이슬란드', '인도네시아', '일본', '룩셈부르크', '멕시코', '뉴질랜드', '']
target_df_1 = df[(df['계정항목'] == '룩셈부르크')].copy()
target_df_2 = df[(df['계정항목'] == '뉴질랜드')].copy()
target_df_3 = df[(df['계정항목'] == '러시아')].copy()


# 3. 보간 적용 (월별 수치 열만)
value_columns = target_df_3.columns[6:]  # 앞의 메타데이터 열 제외
target_df_1[value_columns] = target_df_1[value_columns].interpolate(axis=1, limit_direction='both')
target_df_2[value_columns] = target_df_2[value_columns].interpolate(axis=1, limit_direction='both')
target_df_3[value_columns] = target_df_3[value_columns].interpolate(axis=1, limit_direction='both')

# 4. 나머지 행 추출
other_df = df[~df['계정항목'].isin(['룩셈부르크', '뉴질랜드', '러시아'])]

# 5. 병합 (원래 순서를 유지하고 싶으면 sort=False)
df = pd.concat([other_df, target_df_1, target_df_2, target_df_3], axis=0, ignore_index=True)
```

```python
df.to_csv("../../extra_data/processed/경제 성장 관련/국제 주요국 경제성장률(연도별).csv", index=False)
```
