```python
#  melt로 월 데이터를 세로로 정리
월컬럼 = [f"{i}월" for i in range(1, 13)]
월매핑 = {f"{i}월": i for i in range(1, 13)}
```

```python
df_zs_long = df_zs.melt(id_vars=['차량 구분', '국가명', '연도'], value_vars=월컬럼, var_name='월', value_name='수출량')
df_cp_long = df_cp.melt(id_vars=['차종', '차량 구분', '거래 유형', '연도'], value_vars=월컬럼, var_name='월', value_name='판매량')
df_fp_long = df_fp.melt(id_vars=['공장명(국가)', '공장 코드', '차종', '연도'], value_vars=월컬럼, var_name='월', value_name='공장판매량')
```

```python
# 월 숫자형 변환
for df in [df_zs_long, df_cp_long, df_fp_long]:
    df['월'] = df['월'].map(월매핑)
df_lp['월'] = pd.to_numeric(df_lp['월'], errors='coerce')
```

```python
#  데이터 병합 순서
df_merge1 = pd.merge(df_zs_long, df_cp_long, on=['차량 구분', '연도', '월'], how='outer')
df_merge2 = pd.merge(df_merge1, df_fp_long, on=['차종', '연도', '월'], how='outer')
df_merged = pd.merge(df_merge2, df_lp, on=['차종', '공장명(국가)', '연도', '월'], how='outer')
```

```python
# 수치형 컬럼 결측치 처리 → 0
for col in ['수출량', '판매량', '공장판매량']:
    if col in df_merged.columns:
        df_merged[col] = pd.to_numeric(df_merged[col], errors='coerce').fillna(0)
```

```python
# 국가 수출 컬럼 평균으로 채우기
국가컬럼 = ['Korea', 'U.S.A', 'Canada', 'Mexico', 'Europe', 'Eastern Europe',
         'Latin America', 'Middle East', 'Africa', 'Asia Pacific', 'India', 'China']
for col in 국가컬럼:
    if col in df_merged.columns:
        df_merged[col] = pd.to_numeric(df_merged[col], errors='coerce')
        df_merged[col] = df_merged[col].fillna(df_merged[col].mean())
```

```python
# 범주형 컬럼 결측치 → 최빈값으로 채우기
범주형 = ['차량 구분', '차종', '공장명(국가)', '국가명', '거래 유형', '공장 코드']
for col in 범주형:
    if col in df_merged.columns:
        df_merged[col] = df_merged[col].fillna(df_merged[col].mode()[0])
```

```python
# 차종 키워드 기반 차량단가 설정
def get_unit_price(model):
    if pd.isna(model):
        return 22000000
    model = model.lower()
    if "ev" in model or "전기차" in model or "아이오닉" in model or "ev6" in model:
        return 40000000
    elif "suv" in model or "투싼" in model or "스포티지" in model:
        return 28000000
    else:
        return 22000000  # 기본 세단

df['차량단가'] = df['차종'].apply(get_unit_price)
```

```python
#  차량가격 = 총판매량 × 차량단가
df['차량가격'] = df['총판매량'] * df['차량단가']
```

```python
# 평균구매금액 = 차량가격
df['평균구매금액'] = df['차량가격']
```

```python
#  구매빈도 = 차종별 평균 판매량
df['구매빈도'] = df.groupby('차종')['총판매량'].transform('mean')
```

```python
# 고객유지기간 (단기 기준: 2년 1개월)
df['고객유지기간'] = 2 + (1 / 12)
```

##  고객 생애 가치 분석 컬럼 설명

###  1. `CLTV` (Customer Lifetime Value)
**고객 생애 가치**는 특정 고객이 기업에 기여할 예상 수익을 정량화한 지표입니다.  
수학적 표현:
$$ CLTV = \text{평균구매금액} \times \text{구매빈도} \times \text{고객유지기간} $$

| 구성 요소          | 설명                                                                 |
|-------------------|--------------------------------------------------------------------|
| 평균구매금액       | 차종별 평균 거래 금액 (총판매량 × 차량단가)                         |
| 구매빈도           | 월별 평균 구매 횟수                                                |
| 고객유지기간       | 단기 분석 기준 2.083년 (2년 1개월) 적용                            |

>  CLTV가 높을수록 해당 차종의 전략적 중요도가 높음을 의미

---

###  2. `CLTV_log` (로그 변환 고객 가치)
**데이터 정규화**를 위한 로그 변환 적용:
$$ CLTV_{\text{log}} = \log(1 + CLTV) $$

| 특징              | 설명                                                                 |
|-------------------|--------------------------------------------------------------------|
| 이상치 완화       | 극단적으로 큰 CLTV 값의 영향을 감소                                 |
| 분포 정규화       | 분석 및 시각화에 적합한 형태로 변환                                |
| 0 값 보존         | log(1+x) 변환으로 CLTV=0인 경우에도 안전하게 처리                   |

---

###  3. `고객등급` (Customer Segment)
**RFM 분석 기반** 4단계 분류 체계:

$$ \text{RFM 점수} = R + F + M $$

| 요소         | 설명                          | 가중치 |
|--------------|-------------------------------|--------|
| R (Recency)  | 최근 구매 시점                | 0-4    |
| F (Frequency)| 구매 빈도                     | 0-4    |
| M (Monetary) | 총 구매 금액                  | 0-4    |

#### 등급 분류 기준
| RFM 점수 범위 | 등급       | 특성                          |
|---------------|------------|-------------------------------|
| 0-5           | 🟡 저위험   | 이탈 가능성 높은 고객          |
| 6-8           | 🟠 중위험   | 관심 필요 고객                 |
| 9-10          | 🔴 고위험   | 핵심 관리 대상 고객            |
| 11-12         | 💎 우수고객 | 프리미엄 관리 대상             |

---

##  전략적 활용 가이드
- **세그먼트 비교**: `CLTV_log` 기준 등급별 평균값 분석 → 자원 배분 우선순위 설정
- **타겟 마케팅**: 우수고객 대상 프리미엄 혜택, 고위험 고객 대상 리텐션 캠페인
- **모니터링**: 월별 등급 변화 추적 → 고객 이탈 신호 포착
- **AI 연계**: 고객 등급을 머신러닝 모델 피처로 활용 → 예측 정확도 향상


```python
#  CLTV 계산
df['CLTV'] = df['평균구매금액'] * df['구매빈도'] * df['고객유지기간']
```

```python
#  로그 스케일 CLTV (이상치 완화)
df['CLTV_log'] = np.log1p(df['CLTV'])
```

```python
#  RFM 기반 세분화
df['최근구매연도'] = df.groupby('차종')['연도'].transform('max')
df['구매횟수'] = df.groupby('차종')['연도'].transform('nunique')
df['총구매금액'] = df['차량가격']
```

```python
df['R'] = pd.qcut(df['최근구매연도'], q=4, labels=False, duplicates='drop') + 1
df['F'] = pd.qcut(df['구매횟수'], q=4, labels=False, duplicates='drop') + 1
df['M'] = pd.qcut(df['총구매금액'], q=4, labels=False, duplicates='drop') + 1
df['RFM점수'] = df[['R', 'F', 'M']].astype(int).sum(axis=1)
```

```python
# 고객등급 분류 (RFM 점수 기반)
df['고객등급'] = pd.cut(df['RFM점수'], bins=[0, 5, 8, 10, 12],
                     labels=['저위험', '중위험', '고위험', '우수고객'])
```
