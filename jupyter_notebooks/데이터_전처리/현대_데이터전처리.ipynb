```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sb
```

```python
import os

base_path = "/Users/mi-yeonlim/Documents/Github/mini-project-2/data/processed"

file_list = {
    "df_zs": "현대_지역별수출실적_전처리.csv",
    "df_cp": "현대_차종별판매실적_전처리.csv",
    "df_fp": "현대_해외공장판매실적_전처리.csv",
    "df_lp": "현대_해외현지판매_전처리.CSV"
}

df = {}

for name, file in file_list.items():
    full_path = os.path.join(base_path, file)
    if os.path.exists(full_path):
        df[name] = pd.read_csv(full_path, encoding='utf-8-sig')
        print(f"{name} 불러오기 성공 ✅")
    else:
        print(f"{name} 파일 없음 ❌ → 경로 확인 필요: {full_path}")


```

```python
df_zs = df["df_zs"]
df_cp = df["df_cp"]
df_fp = df["df_fp"]
df_lp = df["df_lp"]
```

```python
#  melt로 월 데이터를 세로로 정리
월컬럼 = [f"{i}월" for i in range(1, 13)]
월매핑 = {f"{i}월": i for i in range(1, 13)}
```

```python
df_zs_long = df_zs.melt(id_vars=['차량 구분', '국가명', '연도'], value_vars=월컬럼, var_name='월', value_name='수출량')
df_cp_long = df_cp.melt(id_vars=['차종', '차량 구분', '거래 유형', '연도'], value_vars=월컬럼, var_name='월', value_name='판매량')
df_fp_long = df_fp.melt(id_vars=['공장명(국가)', '공장 코드', '차종', '연도'], value_vars=월컬럼, var_name='월', value_name='공장판매량')
```

```python
# 월 숫자형 변환
for df in [df_zs_long, df_cp_long, df_fp_long]:
    df['월'] = df['월'].map(월매핑)
df_lp['월'] = pd.to_numeric(df_lp['월'], errors='coerce')
```

```python
#  데이터 병합 순서
df_merge1 = pd.merge(df_zs_long, df_cp_long, on=['차량 구분', '연도', '월'], how='outer')
df_merge2 = pd.merge(df_merge1, df_fp_long, on=['차종', '연도', '월'], how='outer')
df_merged = pd.merge(df_merge2, df_lp, on=['차종', '공장명(국가)', '연도', '월'], how='outer')
```

```python
# 수치형 컬럼 결측치 처리 → 0
for col in ['수출량', '판매량', '공장판매량']:
    if col in df_merged.columns:
        df_merged[col] = pd.to_numeric(df_merged[col], errors='coerce').fillna(0)
```

```python
# 국가 수출 컬럼 평균으로 채우기
국가컬럼 = ['Korea', 'U.S.A', 'Canada', 'Mexico', 'Europe', 'Eastern Europe',
         'Latin America', 'Middle East', 'Africa', 'Asia Pacific', 'India', 'China']
for col in 국가컬럼:
    if col in df_merged.columns:
        df_merged[col] = pd.to_numeric(df_merged[col], errors='coerce')
        df_merged[col] = df_merged[col].fillna(df_merged[col].mean())
```

```python
# 범주형 컬럼 결측치 → 최빈값으로 채우기
범주형 = ['차량 구분', '차종', '공장명(국가)', '국가명', '거래 유형', '공장 코드']
for col in 범주형:
    if col in df_merged.columns:
        df_merged[col] = df_merged[col].fillna(df_merged[col].mode()[0])
```
