```python
import pandas as pd
```

```python
# 엑셀 파일 경로
file_path = "2025년 현지판매실적.xlsx"  # 파일명 수정 필요
```

```python
# 엑셀 파일 열기
xls = pd.read_excel(file_path)

```

```python
xls
```

```python
# 결과 저장용 리스트
combined_data = []

# 반복적으로 각 시트를 변환
for sheet in xls.sheet_names:
    df = pd.read_excel(xls, sheet_name=sheet)
    
    # 유효한 데이터만 필터링 (예: 공장, 차종, 수출국 컬럼이 존재하는 경우)
    df = df.dropna(subset=['공장', '차종', '수출국'], how='all')
    
    # '월' 컬럼 추가 (시트 이름 → YYYY-MM 형식으로 변환 가능하면 좋음)
    df['월'] = sheet  # 필요 시 'sheet'를 '2016-01' 형식으로 수정해도 됨
    
    # Melt (Wide → Long 형태 변환)
    df_melted = df.melt(id_vars=['공장', '차종', '수출국', '월'],
                        var_name='판매월', value_name='수출량')
    
    # 판매월 정보가 날짜로 되어있으면 병합하지 않고 그대로 사용
    combined_data.append(df_melted)

# 모든 월 데이터 하나의 DataFrame으로 결합
final_df = pd.concat(combined_data, ignore_index=True)

# 컬럼 순서 정렬
final_df = final_df[['월', '판매월', '공장', '차종', '수출국', '수출량']]

# 결과 확인
final_df.head()
```
